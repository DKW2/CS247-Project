```python
import matplotlib.pyplot as plt
import numpy as np

models = ("Baseline CNN", "Optimized \nLSTM", "LSTM +\n Data \nAugmentation", "LSTM + \nData Sampling", "Final Model")
model_performance = {
    'Validation CER': (19.30, 14.55, 13.97, 15.06, 14.08),
    'Test CER': (22.17, 16.29, 15.15, 17.35, 15.34),
}

x = np.arange(len(models))  # the label locations
width = 0.25  # the width of the bars
multiplier = 0

fig, ax = plt.subplots(layout='constrained')

for attribute, measurement in model_performance.items():
    offset = width * multiplier
    rects = ax.bar(x + offset, measurement, width, label=attribute)
    ax.bar_label(rects, padding=3)
    multiplier += 1

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('CER')
ax.set_title('Performance of LSTM Models')
ax.set_xticks(x + width, models)
ax.legend(loc='upper right')
ax.set_ylim(0, 25)
plt.savefig( "final_results.png" )

plt.show()
```
